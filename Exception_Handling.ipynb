## Q5)How would you catch and handle division by zero error in Python ?

Ans:To catch and handle a division by zero error in Python, you can use a try-except block. The specific exception for division by zero is ZeroDivisionError.

Explanation:

Try block: Contains the code that might raise a ZeroDivisionError.

Except block: Catches the ZeroDivisionError and prints a friendly error message if division by zero occurs.

Else block: Executes only if no exception occurs, displaying the result.

This ensures that the program doesn't crash and handles the error gracefully.

Here’s an example:

```python
try:
    numerator = 10
    denominator = 0
    result = numerator / denominator
except ZeroDivisionError:
    print("Error: Division by zero is not allowed.")
else:
    print(f"Result: {result}")
```

##Q18) How would you implement a basic logging setup that logs to a file with rotation after 1MB ?

Ans: To implement a basic logging setup in Python that logs to a file with rotation after it reaches 1MB, you can use the logging module along with logging.handlers.RotatingFileHandler. This handler automatically rotates the log file when it exceeds a specified size, and it can keep multiple backup copies.

##Here’s a step-by-step implementation:

Set up a basic logging configuration.

Use RotatingFileHandler to log messages to a file and rotate the log file after it exceeds 1MB.

Configure the maximum number of backup files to retain.


```python
import logging
from logging.handlers import RotatingFileHandler

# Set up the logging configuration
def setup_logging():
    # Create a RotatingFileHandler with a max log file size of 1MB and keep 3 backup files
    handler = RotatingFileHandler('app.log', maxBytes=1_000_000, backupCount=3)
    handler.setLevel(logging.INFO)  # Set the logging level to INFO

    # Create a formatter and set it for the handler
    formatter = logging.Formatter('%(asctime)s - %(name)s - %(levelname)s - %(message)s')
    handler.setFormatter(formatter)

    # Set up the root logger
    logger = logging.getLogger()
    logger.setLevel(logging.INFO)
    logger.addHandler(handler)

    return logger

# Function to demonstrate logging
def demo_logging(logger):
    for i in range(10000):  # Log a lot of messages to generate log file rotation
        logger.info(f"This is log message number {i}")

if __name__ == "__main__":
    logger = setup_logging()  # Set up the logging configuration
    demo_logging(logger)  # Log messages to trigger rotation
    print("Logging is set up and messages are being logged.")
```

##Q21) Write a Python program that reads a file and prints the number of occurrences of a specific word ?

Ans: Here's a Python program that reads a file and counts the number of occurrences of a specific word:

Explanation:

File Reading:

The program opens the file using a context manager (with open(file_name, 'r')) and reads its contents using file.read().

Counting Word Occurrences:

The split() method splits the entire content of the file into words. By default, split() splits by any whitespace (spaces, newlines, etc.).
The lower() method is used to make both the file content and the search word lowercase to ensure the search is case-insensitive.
The count() method counts the occurrences of the specified word in the list of words.

Error Handling:

The program uses a try-except block to catch errors like FileNotFoundError if the file does not exist.
A generic Exception is also used to catch any other unexpected errors.


```python
def count_word_occurrences(file_name, word_to_count):
    try:
        # Open the file using a context manager
        with open(file_name, 'r') as file:
            # Read the file contents
            content = file.read()

            # Count occurrences of the specific word
            word_count = content.lower().split().count(word_to_count.lower())

        print(f"The word '{word_to_count}' appears {word_count} times in the file.")

    except FileNotFoundError:
        print(f"Error: The file '{file_name}' was not found.")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")

# Call the function
file_name = 'example.txt'  # Replace with your file name
word_to_count = 'python'   # Replace with the word you want to count
count_word_occurrences(file_name, word_to_count)
```

##Q22) How can you check if a file is empty before attempting to read its contents ?

Ans: In Python, you can check if a file is empty before attempting to read its contents by examining the file's size using os.stat() or checking the file's contents directly. If the file size is 0, it is considered empty. Here's how you can do it:

Approach 1: Using os.stat() to Check File Size

You can use the os module to get the file's statistics, including its size, and check if it's 0 bytes.



```python
import os

def is_file_empty(file_name):
    # Get file statistics
    file_stats = os.stat(file_name)

    # Check if the file size is 0 bytes
    return file_stats.st_size == 0

def read_file_if_not_empty(file_name):
    # Check if the file is empty
    if is_file_empty(file_name):
        print(f"The file '{file_name}' is empty.")
        return

    # Read the file contents if it's not empty
    with open(file_name, 'r') as file:
        content = file.read()
        print(f"File contents:\n{content}")

# Example usage
file_name = 'example.txt'  # Replace with your file name
read_file_if_not_empty(file_name)
```

##Explanation:

os.stat(file_name):

This function returns the status of the file. st_size contains the size of the file in bytes.
st_size == 0:

This checks if the file size is 0 bytes, which indicates that the file is empty.
read_file_if_not_empty(file_name):

This function first checks if the file is empty by calling is_file_empty(file_name). If the file is empty, it prints a message. If not, it reads and prints the file's contents.

Approach 2: Using file.read() and Checking for Empty String

Another way is to try to read the file and check if the returned content is an empty string:

```python
def read_file_if_not_empty(file_name):
    # Open the file and check if it's empty
    with open(file_name, 'r') as file:
        content = file.read()

        if not content:  # If content is an empty string
            print(f"The file '{file_name}' is empty.")
        else:
            print(f"File contents:\n{content}")

# Example usage
file_name = 'example.txt'  # Replace with your file name
read_file_if_not_empty(file_name)
```

##Q23) Write a Python program that writes to a log file when an error occurs during file handling.

Ans:To write a Python program that logs errors during file handling, you can use Python's built-in logging module. The program will handle errors (such as FileNotFoundError, PermissionError, etc.) and log them into a log file.




```python
import logging

# Set up logging to log errors to a file
def setup_logging():
    logging.basicConfig(
        filename='file_error_log.txt',  # Log file where errors will be saved
        level=logging.ERROR,            # Log only ERROR and higher levels
        format='%(asctime)s - %(levelname)s - %(message)s',  # Log format
    )

# Function to demonstrate file handling with error logging
def handle_file_operations():
    try:
        # Try opening a file and performing some operations
        with open('non_existent_file.txt', 'r') as file:
            content = file.read()
            print(content)

    except FileNotFoundError as e:
        # Log the error if the file is not found
        logging.error(f"File not found: {e}")
        print("An error occurred while trying to open the file. Check the log for details.")

    except PermissionError as e:
        # Log the error if there is a permission issue
        logging.error(f"Permission denied: {e}")
        print("You don't have permission to access this file. Check the log for details.")

    except Exception as e:
        # Log any other unexpected errors
        logging.error(f"An unexpected error occurred: {e}")
        print("An unexpected error occurred. Check the log for details.")

# Set up logging and handle file operations
setup_logging()
handle_file_operations()
```
